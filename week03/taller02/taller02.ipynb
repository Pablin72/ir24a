```python
import os
import codecs
import re

directorio = "textos"

def limpiar_texto(texto):
    # Eliminar caracteres no deseados
    texto = re.sub(r"[,.;?!\"-]", "", texto)
    texto = texto.replace('"','').replace(':','')
    # Opcional: Convertir a minúsculas
    texto = texto.lower()
    # Opcional: Eliminar espacios en blanco adicionales
    texto = re.sub(r"\s+", " ", texto)
    return texto

# Obtener la lista de archivos en el directorio
archivos_txt = [archivo for archivo in os.listdir(directorio)]

for archivo in archivos_txt:
    ruta_archivo = os.path.join(directorio, archivo)
    with codecs.open(ruta_archivo, 'r') as f:
        contenido = f.read()
    contenido_limpio = limpiar_texto(contenido)
    with open(ruta_archivo, 'w') as f:
        f.write(contenido_limpio)
```

```python
import pandas as pd
from collections import defaultdict

# Define the path to the directory containing the text files
CORPUS_DIR = directorio
documents = {}

# Dictionary to store word counts for each document
word_counts = defaultdict(lambda: defaultdict(int))

# Iterate through each file in the directory
for filename in os.listdir(CORPUS_DIR):
    if filename.endswith('.txt'):
        file_path = os.path.join(CORPUS_DIR, filename)
        with open(file_path, 'r') as file:
            document_text = file.read().lower()  # Read and convert to lowercase
            documents[filename] = document_text

            # Count word occurrences in the current document
            words = document_text.split()
            for word in words:
                word_counts[word][filename] += 1

# Create a DataFrame from the word counts
df = pd.DataFrame(word_counts)

# Fill NaN values with 0 (for words not present in some documents)
df.fillna(0, inplace=True)

# Normalize the DataFrame
df_normalized = df.div(df.max(axis=1), axis=0)

df_normalized

```

```python
df_normalized.to_csv("corpus_data.csv", index=False)  # Save without index column
```

```python
# Define the path to the directory containing the text files
QUERIES_DIR = "queries"
documents = {}

# Dictionary to store word counts for each document
queries = defaultdict(lambda: defaultdict(int))

# Iterate through each file in the directory
for filename in os.listdir(QUERIES_DIR):
    if filename.endswith('.txt'):
        file_path = os.path.join(QUERIES_DIR, filename)
        with open(file_path, 'r') as file:
            document_text = file.read().lower()  # Read and convert to lowercase
            documents[filename] = document_text

            # Count word occurrences in the current document
            words = document_text.split()
            for word in words:
                queries[word][filename] += 1

# Create a DataFrame from the word counts
df_queries = pd.DataFrame(queries)

# Fill NaN values with 0 (for words not present in some documents)
df_queries.fillna(0, inplace=True)

# Normalize the DataFrame
df_queries_normalized = df_queries.div(df_queries.max(axis=1), axis=0)

df_queries_normalized
```

```python
df_queries_normalized.to_csv("queries_data.csv", index=False)
```
